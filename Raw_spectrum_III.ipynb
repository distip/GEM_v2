```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Raw_spectrum_merged')
```

```python
ASD_1 = data.loc[data['ASD']==1 , :]
```

```python
ASD_1_plt = ASD_1.iloc[: , 10:].T
```

```python
ASD_2= data.loc[data['ASD']==2 , :]
```

```python
ASD_2_plt = ASD_2.iloc[: , 10:].T
```

```python
plt.plot(ASD_1_plt, color= 'r')
plt.plot(ASD_2_plt, color= 'g')
plt.legend()
plt.xticks(ticks=['450', '700', '900', '1500', '2000', '2500'])
plt.xlabel('Wavelength')
plt.ylabel('Reflectance')
```
