```python
ASD_1 = pd.read_csv('ASD_1', low_memory = False)
ASD_2 = pd.read_csv('ASD_2', low_memory= False)
```

```python
ASD_1_cals = []
for i in ASD_1['PLOT ID']:
    if '-C' in i:
        ASD_1_cals.append(i)
        
            
```

```python
ASD_2_cals = []
for i in ASD_2['PLOT ID']:
    if '-C' in i:
        ASD_2_cals.append(i)
```

```python
ASD_2_cals
```

```python
ASD_1_cal_spec = ASD_1.loc[ASD_1['PLOT ID'].isin(ASD_1_cals) , :]
```

```python
ASD_2_cal_spec = ASD_2.loc[ASD_2['PLOT ID'].isin(ASD_2_cals) , :]
```

```python
ASD_1_cal_spec.shape
```

```python
ASD_2_cal_spec.shape
```

```python
ASD_2_cal_spec
```

```python
calibration = pd.concat([ASD_1_cal_spec, ASD_2_cal_spec])
```

```python
calibration.loc[calibration['PLOT ID'] == '1233-C']
```

```python
calibration.head()
```

```python
ids = []
for i in calibration['PLOT ID']:
    a=i[0:4]
    ids.append(a)
```

```python
calibration.drop('PLOT ID', axis=1, inplace= True)
```

```python
calibration.insert(1, 'PLOT ID', ids)
```

```python
calibration.drop('Unnamed: 0', axis=1 , inplace=True)
```

```python
calibration.head(20)
```

```python
cal1 = calibration.loc[calibration['ASD'] == 1, :].groupby('PLOT ID', as_index = False).mean()
```

```python
cal2 = calibration.loc[calibration['ASD'] == 2, :].groupby('PLOT ID', as_index=False).mean()
```

```python
calibration_v2= pd.concat([cal1, cal2], join = 'outer')
```

```python
calibration_v2.head()
```

```python
calibration_v2.to_csv('calibration')
```
