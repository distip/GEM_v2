```python
import pandas as pd
import numpy as np
```

```python
spectrum = pd.read_csv('Raw_spectrum_merged.csv')
```

```python
gt = pd.read_csv('BGEM_2022_gt.csv')
```

```python
data = pd.merge(gt, spectrum, how='outer')
```

```python
data
```

```python
data.drop(data.filter(regex='Unname'), axis=1 , inplace=True)
```

```python
data =data.sort_values('PLOT ID')
```

```python
data.reset_index(inplace=True, drop=True)
```

```python
data
```

```python
CHL= data[['CHL1', 'CHL2', 'CHL3']].mean(axis=1) 
```

```python
data.insert(3, 'CHL', CHL)
```

```python
SLA = data['Leaf_area']/data['Dry_weight']
```

```python
data.insert(3, 'SLA', SLA)
```

```python
LWC = (data['Fresh_weight']-data['Dry_weight']) / data['Fresh_weight'] * 100
```

```python
data.insert(3, 'LWC' , LWC)
```

```python
data.to_csv('Spectrum_ground_truth_v1.csv', index=False)
```
